```python
from pytrends. request import TrendReq
import pandas as pd
```

```python
# Configuração inicial do PyTrends
pytrends = TrendReq(hl='pt-BR', tz=360)

# Definição da palavra-chave para a pesquisa
keywords = ['sustentabilidade']

# Configuração do período (últimos 5 anos)
from datetime import datetime, timedelta
end_date = datetime. today() . strftime( '%Y-%m-%d' )
start_date = (datetime.today() - timedelta(days=5*365) ) . strftime( '%Y-%m-%d' )
timeframe = f'{start_date} {end_date}'

# Busca pelo interesse ao longo do tempo
pytrends.build_payload(kw_list=keywords, timeframe=timeframe)
interest_over_time_df = pytrends.interest_over_time()

# Exibindo os resultados
print(interest_over_time_df)
```

```python
pytrends.categories()
```

```python
df = pytrends.interest_by_region().reset_index()
df.sort_values(keywords, ascending=False, inplace=True)
df.head(10)
```

```python
pytrends.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=False)
```

```python
kw = pytrends.suggestions(keyword='keywords')
df =pd.DataFrame(kw)
df.drop(columns='mid')
```
